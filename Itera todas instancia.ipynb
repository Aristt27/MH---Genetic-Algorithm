```python
### Imports
from Import_csv_instance import instancia_csv
from GA_Genetic_algorithm import Genetic_Algorithm
from Solution_visualizer import solution_visualize
from Evolution_statistical_analysis import evolution_statistical_analysis
from Genetic_iterator import genetic_iterator
```

```python
CS = ['Tournament', 'Mix', 'Ranking']
CT = ['ONE_CUT', 'MULTI_CUT']
tsizes = [2,4,6]
iters = 10
for i in range(1,5):
    filepath = 'results/'
    
    instancia = instancia_csv("instances/i"+str(i)+".csv")
    Rooms       = 2
    Instancia = (instancia, Rooms)
    ## Configuração da população inicial, 
    greedy_pop   = 20 # greedy -> resultado do guloso.
    random_pop   = 20 # sample aleatório.

    pop_inicial  = (greedy_pop, random_pop)
    # Cortes por geração 
    elite_cut    = 2   # elite são os melhores fitness
    lucky_cut    = 12 # lucky são os outros

    gen_cuts     = (elite_cut, lucky_cut)
    
    ## Parametros da mutação
    β = 0.05 # Probabilidade de escolher tal filho para mutar;
    mutation_params = (β)

    ## Criterios de parada:
    generations = 1000    # Max_iter

    stop_len = 10 # quantas iterações seguidas
    K        = 11 # a media das K melhores 
    tol      = 1  # tem q estar a uma distancia tol do melhor
    
    α            = 0.9 
    for Cross_selection in CS:
        for cut_type in CT:
            if Cross_selection != 'Ranking':
                for tournament_size in tsizes:


                        filename = filepath+'ri' + str(i) + '_'+str(tournament_size)+Cross_selection[0]+cut_type[0]+str(iters)+'.txt'

                        cross_params = (Cross_selection, α, cut_type, tournament_size)

                        set_params = [pop_inicial, gen_cuts, cross_params, mutation_params]
                        set_stop_criteria = [generations, stop_len, K, tol]

                        asd = genetic_iterator(Instancia, set_params, set_stop_criteria, 10, output=filename)
            else:
                tournament_size = 1
                filename = filepath+'ri' + str(i) + '_'+str(tournament_size)+Cross_selection[0]+cut_type[0]+str(iters)+'.txt'

                cross_params = (Cross_selection, α, cut_type, tournament_size)

                set_params = [pop_inicial, gen_cuts, cross_params, mutation_params]
                set_stop_criteria = [generations, stop_len, K, tol]

                asd = genetic_iterator(Instancia, set_params, set_stop_criteria, 10, output=filename)
```
